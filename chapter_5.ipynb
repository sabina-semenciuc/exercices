## Stack

```python
class Stack:  # Creating the Stack class

  def __init__(self):
    self.stack = []  # Initialize an empty list to store the items in the stack
  
  def is_empty(self):
    # Check if the stack is empty
    # Returns: True if the length of the stack list is 0 (empty), otherwise False
    return True if len(self.stack) == 0 else False

  def push(self, item):
    # Push an item onto the stack
    # Args:
    #     item: The item to be pushed onto the stack
    self.stack.append(item)

  def pop(self):
    # Pop and return an item from the top of the stack
    # Returns: The item removed from the top of the stack if the stack is not empty, otherwise None
    return None if self.is_empty() else self.stack.pop()
```

```python
def check(expression):
    # Function to check if the opening and closing brackets in the expression are balanced
    opening = "[{("  # Opening bracket characters
    closing = "]})"  # Closing bracket characters
    stack = []  # Stack to store opening brackets encountered

    for character in expression:
        if character in opening:
            stack.append(character)  # Push opening brackets onto the stack
        elif character in closing:
            if len(stack) == 0:
                return False  # Unbalanced closing bracket found
            elif opening.index(stack.pop()) != closing.index(character):
                return False  # Mismatched opening and closing brackets found
    return len(stack) == 0  # Return True if all opening brackets are matched and stack is empty
```

## Queue

```python
class Queue:
    # Class representing a Queue data structure

    def __init__(self):
        # Initialize an empty list to store the items in the queue
        self.queue = []

    def is_empty(self):
        # Check if the queue is empty
        # Returns: True if the length of the queue list is 0 (empty), otherwise False
        return True if len(self.queue) == 0 else False

    def enqueue(self, item):
        # Add an item to the rear of the queue
        # Args:
        #     item: The item to be added to the queue
        self.queue.append(item)

    def dequeue(self):
        # Remove and return an item from the front of the queue
        # Returns: The item removed from the front of the queue if the queue is not empty, otherwise None
        return None if self.is_empty() else self.queue.pop(0)
```

### 5. Write the output of the codes from the example that uses a queue.

```python
queue = Queue()  # Create an instance of the Queue class
items = [10 * i for i in range(1, 11)]  # Create a list of items ranging from 10 to 100 with a step of 10

for item in items:
  queue.enqueue(item)  # Add each item to the queue

  # Check if the tens place of the item divided by 10 is even
  if (item // 10) % 2 == 0:
    queue.dequeue()  # If the condition is met, remove an item from the queue

print(queue.queue)  # Print the final queue after the operations
```

```python
class Deque:
  def __init__(self): 
    self.queue = []  # Initialize an empty list to store the items in the deque

  def add_first(self, item):
    ''' Add an item to the front of the deque'''
    self.queue.insert(0, item)  # Insert the item at the beginning of the list

  def remove_first(self):
    '''Remove and return the item from the front of the deque'''
    if not self.is_empty():  # Check if the deque is not empty
      return self.queue.pop(0)  # Remove and return the item at index 0
    else:
      return None  # Return None if the deque is empty

  def add_last(self, item):
    '''Add an item to the rear of the deque'''
    self.queue.append(item)  # Append the item at the end of the list

  def remove_last(self):
    ''' Remove and return the item from the rear of the deque'''
    if not self.is_empty():  # Check if the deque is not empty
      return self.queue.pop()  # Remove and return the last item in the list
    else:
      return None  # Return None if the deque is empty

  def is_empty(self):
    '''Check if the deque is empty'''
    return len(self.queue) == 0  # Return True if the deque is empty, False otherwise

  def size(self):
    '''Return the size of the deque'''
    return len(self.queue)  # Return the length of the list, which represents the size of the deque
```
