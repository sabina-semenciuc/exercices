### 1. Receive the coordinates (x1, y1), (x2, y2) of two points from the user and print the distance between the two points as follows. To do this, implement the distance(x1, y1, x2, y2) function.
Coding guideline: Refer to the figure below for the formula to find the distance between two points.

```python
while True:
    try:
        x1 = int(input("Enter the x1 coordinate:"))  # Prompt user to input x1 coordinate
        y1 = int(input("Enter the y1 coordinate:"))  # Prompt user to input y1 coordinate
        x2 = int(input("Enter the x2 coordinate:"))  # Prompt user to input x2 coordinate
        y2 = int(input("Enter the y2 coordinate:"))  # Prompt user to input y2 coordinate
        break  # If input is successful, exit the loop
    except ValueError:
        print("Values inserted incorrectly. Try again!")  # If input is not an integer, catch the ValueError and print an error message

d = int(((y2 - y1)**2 + (x2 - x1)**2)**(1/2))  # Calculate the distance between the two points using the distance formula
print(f"The distance between the two points given is: {d}")  # Print the calculated distance
```

### 2. A palindrome is a sentence, word, or string that is the same as reading any character from back to front. Words like level, eve, and radar are examples of the palindrome. Let's use a recursive call to determine a palindrome. Define a function called is_palindrome and write a program that receives a string from the user and prints whether the string is a palindrome or not.
Coding guideline: Call the is_palindrome function inside the is_palindrome function.

```python
def is_palindrome(string):
    # Base case: If the string is empty or contains only one character, it is a palindrome
    if len(string) <= 1:
        return True
    
    # Recursive case: Check if the first and last characters are equal
    if string[0] == string[-1]:
        # Remove the first and last characters and check if the remaining string is a palindrome
        return is_palindrome(string[1:-1])
    
    return False  # If the first and last characters are not equal, the string is not a palindrome

# Get input from the user
user_input = input("Enter a string: ")

# Check if the string is a palindrome
if is_palindrome(user_input):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

### 5. Implement multiplication (*) and division (/) of two vectors using the mul and truediv methods learned in unit 21. Assuming that v1 is (30, 40) and v2 is (10, 20), write a code that prints the output as below as the result of the multiplication and division of two vectors.
Coding guideline: Write a code that extends the functionality of the Vector2D class you learned in this chapter.


```python
v1 = (30, 40)
v2 = (10, 20)

result_mul = tuple(map(lambda x, y: x * y, v1, v2))  # Multiplication of two vectors
result_div = tuple(map(lambda x, y: x / y, v1, v2))  # Division of two vectors

print(f"Multiplication of two vectors: {v1} * {v2} = {result_mul}")

print(f"\nDivision of two vectors: {v1} / {v2} = {result_div}")
```

```python
import operator

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __mul__(self, other):
        return Vector(operator.mul(self.x, other.x), operator.mul(self.y, other.y))
    
    def __truediv__(self, other):
        return Vector(operator.truediv(self.x, other.x), operator.truediv(self.y, other.y))
    
    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(30, 40)
v2 = Vector(10, 20)

result_mul = v1 * v2  # Multiplication of two vectors
result_div = v1 / v2  # Division of two vectors

print("Multiplication of two vectors:")
print(f"v1 * v2 = {result_mul}")

print("\nDivision of two vectors:")
print(f"v1 / v2 = {result_div}")
```
